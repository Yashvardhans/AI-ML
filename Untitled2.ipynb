```python
import os
from sklearn.model_selection import train_test_split


dataset_root = '/content/dataset/Garbage classification'

file_paths = []
labels = []


categories = os.listdir(dataset_root)
for category in categories:
    category_path = os.path.join(dataset_root, category)
    if os.path.isdir(category_path):
        for filename in os.listdir(category_path):
            file_path = os.path.join(category_path, filename)
            file_paths.append(file_path)
            labels.append(category)


test_size = 0.2
random_state = 42

X_train, X_test, y_train, y_test = train_test_split(
    file_paths, labels, test_size=test_size, random_state=random_state
)


```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle


def load_images(file_paths, labels):
    images = []
    for file_path in file_paths:
        img = load_img(file_path, target_size=(150, 150))
        img_array = img_to_array(img)
        images.append(img_array)
    images = np.array(images)


    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)

    return images, encoded_labels


X_train_images, y_train_encoded = load_images(X_train, y_train)
X_test_images, y_test_encoded = load_images(X_test, y_test)


X_train_images, y_train_encoded = shuffle(X_train_images, y_train_encoded, random_state=42)


X_train_images = X_train_images / 255.0
X_test_images = X_test_images / 255.0

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')
])


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


model.fit(X_train_images, y_train_encoded, epochs=10, batch_size=32, validation_data=(X_test_images, y_test_encoded))
```

```python

predictions = model.predict(X_test_images)


predicted_labels = (predictions > 0.5).astype(int).reshape(-1)


for i in range(10):
    print(f"Predicted: {predicted_labels[i]}, True Label: {y_test_encoded[i]}")
```
